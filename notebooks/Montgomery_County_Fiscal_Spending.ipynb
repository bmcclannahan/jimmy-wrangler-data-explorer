Aggregate Overdose data by age range and gender and substance

```python
agg = {'Count': 'sum',}
df = df_overdose.groupby(['age_group','PCR_Patient_Gender','Substance']).aggregate(agg)
df
```

Merge homeless data with overdose data

```python
location = [200]*74
df['Homeless per 100,000 Individuals'] = location
df
```

Although at first glance the merged data is seemingly unimportant, this dataset could be very valuable when compared with similar data from other cities. By adding the homeless data to the overdose data, one could compare the homeless rates of other cities to the homeless rates of Austin with the overdose rates of different demographics of other cities to Austin. This would allow one to see how homelessness affects different demographics and their likelihood to abuse drugs.
